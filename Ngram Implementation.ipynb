### Data Cleaning and Preprocessing

```python
import nltk
import re
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()
lemma = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i]) #regex expression
    review = review.lower() #will do the lowercase
    review = review.split() #will split the sentence
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')] #portor stemming on words that are not stopwords
    review = ' '.join(review) #join the words back seperated by space
    corpus.append(review) #append it back in a list
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i]) #regex expression
    review = review.lower() #will do the lowercase
    review = review.split() #will split the sentence
    review = [lemma.lemmatize(word) for word in review if not word in stopwords.words('english')] #lemmatizing on words that are not stopwords
    review = ' '.join(review) #join the words back seperated by space
    corpus.append(review) #append it back in a list
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500,binary = True)
X = cv.fit_transform(corpus).toarray()
```
